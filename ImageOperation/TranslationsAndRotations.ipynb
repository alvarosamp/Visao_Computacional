## Translations

This an affine transofrm that simpy shifts the position oftan image

We use cv2.warpAffine to implement these transformations

cv2.warpAffine(image, T, width, height)

Tx -> Represents the shift along the x-axis
Ty -> Represents the shift along the y-axis

```python
# Display the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.show()

# Get the dimensions of the image
height, width = image.shape[:2]

# Calculate the translation values
quarter_height, quarter_width = height / 4, width / 4

# Create the translation matrix
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

# Apply the translation to the image
img_translation = cv2.warpAffine(image, T, (width, height))

# Display the translated image
plt.imshow(cv2.cvtColor(img_translation, cv2.COLOR_BGR2RGB))
plt.title('Translation')
plt.show()
```

## Rotations 
cv2.getRotationMatrix2D(rotation_center_x, rotation_center_y, angle of rotation, scale)

```python
# Load the input image
image = cv2.imread(r'C:\Users\vish8\OneDrive\Desktop\Cursos\ComputerVision\OpenCv\ImageOperation\extracted_images\images\Volleyball.jpeg')
height, width = image.shape[:2]

# Divide by two to rotate the image around its center
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 0.5)
# Apply the rotation matrix to the image
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

# Display the rotated image
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated 90 degrees')
plt.show()
```

## Rotations with cv2.transpose

```python
rotated_image = cv2.transpose(image)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Transpose')
plt.show()
```

```python
rotated_image = cv2.flip(image, 1)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Flipped')
plt.show()
```
