```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

def imshow(title = 'title', image = None, size = 10):
    w,h = image.shape[0],image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize =(size * aspect_ratio.size))
    plt.imshow(image, cmap = 'gray')
    plt.title(title)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Making a blank square canvas
image = np.zeros((512, 512, 3)).astype(np.uint8)
image_gray = np.zeros((512, 512)).astype(np.uint8)

# Black would be the same as a grayscale or color image (same for white)
plt.imshow(image)
plt.title('Black Square')
plt.show()

plt.imshow(image_gray, cmap='gray')
plt.title('Black Square Gray')
plt.show()
```

### Draw a line over our black square

```python
#cv2.line(image,starting cordinates, ending cordinates, color, thickness)
image = cv2.line(image,(0,0), (511,511), (255,127,0), 5)
plt.imshow(image)
plt.title('Blue Line')
plt.show()
```

## Drawing rectangles

```python
image = np.zeros((512, 512, 3)).astype(np.uint8)

#Thickness  - If positive, it means that the thickness is measured in pixels. If negative, it means that the thickness is measured in pixels and the line is filled.
image = cv2.rectangle(image, (100,100), (300,250), (127,50,127), 10)        
plt.imshow(image)
plt.title('rectangle')
plt.show()
```

## Now let's draw some circles 
cv2.circles(image, center, radius, color, fill)

```python
image = np.zeros((512,512,3), np.uint8)
cv2.circle(image, (350,350), 100, (15,75,50), -1)
plt.imshow(image)
plt.title('circle')
plt.show()
```

## Polygons
cv2.polylines(image, points, Closed?, color, thickness)

```python
image = np.zeros((512,512,3), np.uint8)

#Define four points
pts = np.array([[10,50], [400,50], [90,200], [50,500]], np.int32)

pts = pts.reshape((-1,1,2))
image  = cv2.polylines(image, [pts], isClosed = True, color = (255,0,255), thickness = 3)
plt.imshow(image)
plt.title('polygon')
plt.show()
```
