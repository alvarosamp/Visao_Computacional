```python
import cv2
image = cv2.imread(r'C:\Users\vish8\OneDrive\Desktop\Cursos\ComputerVision\OpenCv\ImageOperation\extracted_images\images\flowers.jpeg')
```

```python
import zipfile
import os

# Caminho para o arquivo ZIP
zip_path = "images.zip"

# Pasta de destino para extrair os arquivos
extract_to = "extracted_images"

# Cria a pasta de destino, se não existir
os.makedirs(extract_to, exist_ok=True)

# Abre o arquivo ZIP e extrai todo o conteúdo
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print(f"Arquivos extraídos para a pasta: {extract_to}")
```

```python
# Displaying imagesa
from matplotlib import pyplot as plt

# Show the image with matplotlib
if image is not None:
	plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
	plt.show()
else:
	print("Image not loaded. Please check the file path.")
```

```python
def imshow(title = '',image = None):
    plt.figure(figsize = (10,10))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
imshow('Flowers',image)
```

```python
## Saving images
cv2.imwrite('new_image.jpg',image)
```
