```python
import cv2
import matplotlib.pyplot as plt
```

```python
#define imshow funciton
def imshow(title = 'Image', image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
#Load the input image
image = cv2.imread(r'C:\Users\vish8\OneDrive\Desktop\Cursos\ComputerVision\OpenCv\ImageOperation\extracted_images\images\castara.jpeg')
imshow('Original image', image)
```

```python
'''
Wee use cvtColor, to convert to grayscale
Its takes 2 arguments, the first being the input image
the second being the color space conversion code
'''

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
imshow('Gray Image', gray_image)

```
